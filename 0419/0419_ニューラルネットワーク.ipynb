```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# x、y座標
X = np.arange(-1.0, 1.0, 0.2)  # 要素数は10個
Y = np.arange(-1.0, 1.0, 0.2)

# 出力を格納する10x10のグリッド
Z = np.zeros((10,10))

# 重み
# 中間層(入力→中間層への重み) 
w_im = np.array([[0.2, 0.3, 0.4, 0.5],
                 [0.4, 0.5, 0.8, 0.3]])
# 出力層(中間→出力層への重み)
w_mo = np.array([[0.1], [0.2], [0.3], [0.4]])
# バイアス
# 中間層
b_im = np.array([-0.1, 0.2, -0.3, 0.4])
# 出力層 
b_mo = np.array([0.3])             

# 中間層(シグモイド関数)
def middle_layer(x, w, b):
    # 実装を追加
    u = np.dot(x, w) + b
    return 1/(1+np.exp(-u))

# 出力層(恒等関数)
def output_layer(x, w, b):
    # 実装を追加
    u = np.dot(x, w) + b
    return u

# グリッドの各マスでニューラルネットワークの演算
for i in range(10):
    for j in range(10):
        
        # 順伝播
        inp = np.array([X[i], Y[j]])        # 入力層
        mid = middle_layer(inp, w_im, b_im) # 中間層
        out = output_layer(mid, w_mo, b_mo) # 出力層
        
        # グリッドにNNの出力を格納
        Z[j][i] = out[0]

# グリッドの表示
plt.imshow(Z, "gray", vmin = 0.0, vmax = 1.0)
plt.colorbar()  
plt.show()
```
